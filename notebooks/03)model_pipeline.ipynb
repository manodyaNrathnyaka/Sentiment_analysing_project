```python
import numpy as np 
import pandas as pd
```

```python
text='great product, i like it.'
```

```python
import re
import string
```

```python
with open('../Static/model/corpora/stopwords/english','r') as file:
    sw=file.read().splitlines()
```

```python
from nltk.stem import PorterStemmer
ps=PorterStemmer()
```

```python
def preprocesing(text):
    data=pd.DataFrame([text],columns=['tweet'])
    
    data["tweet"]=data["tweet"].apply(lambda x:" ".join(x.lower() for x in x.split()))
    data["tweet"] = data["tweet"].apply(lambda x: re.sub(r'http\S+|www\S+', ' ', x))
    data['tweet']=data["tweet"].apply(lambda x: re.sub(r"[^\w\s]"," ",x))
    data["tweet"]=data["tweet"].str.replace(r'\d+','',regex=True)
    data["tweet"]=data['tweet'].apply(lambda x: ' '.join(x for x in x.split() if x not in sw))
    data["tweet"]=data["tweet"].apply(lambda x: " ".join(ps.stem(x) for x in x.split()))
    return data["tweet"]
```

```python
preprocessed_text=preprocesing(text)
```

```python
preprocessed_text
```

```python
vocab=pd.read_csv('../static/model/vocabulary.txt', header=None)
token=vocab[0].tolist()
```

```python
token
```

```python
def vectorization(ds,vocabulary):
    vectorized_list=[]
    for sentence in ds:
        sent_list=np.zeros(len(vocabulary))
        
        for i in range(len(vocabulary)):
            if vocabulary[i] in sentence.split():
                sent_list[i]=1

        vectorized_list.append(sent_list)
        
    return np.asarray(vectorized_list)
    
```

```python
vectorized_txt=vectorization(preprocessed_text,token)
```
